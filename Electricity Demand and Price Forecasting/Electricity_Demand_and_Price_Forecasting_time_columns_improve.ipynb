```python
from sklearn.decomposition import PCA
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

def apply_PCA(X_input, cum_variance, if_apply): 
    if if_apply:   
        pca = PCA(n_components = cum_variance)
        # make pipeline to first standardize then apply PCA on data
        scaler_pca = make_pipeline(MinMaxScaler(), pca)
        X_pca = scaler_pca.fit(X_input).transform(X_input)
        return X_pca    
    else:        
        return np.array(X_input)
```

```python
params_pca = {'cum_variance' : 0.8, 'if_apply' : True }
X_pca = apply_PCA(X, **params_pca)
X_pca.shape
```

```python
def windowing(X_input,y_input, history_size):
    data = []
    labels = []
    for i in range(history_size, len(y_input)):
        data.append(X_input[i - history_size : i, :])
        labels.append(y_input[i])
        
    return np.array(data), np.array(labels).reshape(-1,1)
```

```python
train_cutoff = int(0.8*X_pca.shape[0])
val_cutoff   = int(0.9*X_pca.shape[0])
scaler_y = MinMaxScaler()
scaler_y.fit(y[:train_cutoff])
y_norm = scaler_y.transform(y)
```

```python
hist_size= 24
data_norm = np.concatenate((X_pca,y_norm), axis = 1)
X_train, y_train = windowing(data_norm[:train_cutoff,:],data_norm[:train_cutoff,-1], hist_size)
X_val, y_val     = windowing(data_norm[train_cutoff :val_cutoff,:],data_norm[train_cutoff:val_cutoff,-1], hist_size)
X_test, y_test   = windowing(data_norm[val_cutoff :,:],data_norm[val_cutoff:,-1], hist_size)
```

```python
def reshaping(X):
    reshaped_x = X.reshape(-1 , X.shape[1] * X.shape[2])
    return reshaped_x
X_train_xgb = reshaping(X_train)
X_val_xgb   = reshaping(X_val)
X_test_xgb  = reshaping(X_test)
```

```python
def plot_results(y_pred_actual, y_test_inv, model, model_name):
    fig, ax = plt.subplots(2, 1, figsize=(12, 6))
    
    ax[0].plot(y_pred_actual[:1000])
    ax[0].plot(y_test_inv[:1000])
    ax[0].legend(['prediction', 'actual'], loc='upper left')
    ax[0].set_title(f'Prediction vs actual price for 1000 observation in test set ({model_name})')
    ax[0].set_xlabel('Observation')
    ax[0].set_ylabel('Price')
    
    if model != 'hybrid':
    
        if model_name == 'XGBoost':

            train_mae = model.evals_result()['validation_0']['mae']
            val_mae = model.evals_result()['validation_1']['mae']
            ax[1].plot(train_mae, label='Training MAE')
            ax[1].plot(val_mae, label='Validation MAE')

        else:

            ax[1].plot(history.history['loss'], label='Training Loss')
            ax[1].plot(history.history['val_loss'], label='Validation Loss')

        ax[1].legend()
        ax[1].set_title(f'Training and validation MAE ({model_name})')
        ax[1].set_xlabel('Iteration/Epochs')
        ax[1].set_ylabel('MAE')    
        
    fig.tight_layout()
    plt.show()
    
```

#### XGBoost Evaluation by Expanding Window Cross Validation

```python
import numpy as np
import xgboost as xgb
from sklearn.metrics import mean_absolute_error

# ✅ Set Expanding Window Parameters
initial_train_size = 1000   # Initial training set size
test_size = 1000            # Each test set size
n_splits = 27               # Expanding window 27 times

mae_scores = []
best_mae = float("inf")  # Store the best MAE
best_model_path = "best_xgb_model.json"  # Path to save best model

for i in range(n_splits):
    # ✅ Define expanding train/test indices
    train_end = initial_train_size + i * test_size
    test_start = train_end
    test_end = test_start + test_size

    # ✅ Ensure test set does not exceed data range
    if test_end > len(X_train_xgb):
        break

    # ✅ Extract train and test data
    X_train_cv, X_test_cv = X_train_xgb[:train_end], X_train_xgb[test_start:test_end]
    y_train_cv, y_test_cv = y_train[:train_end], y_train[test_start:test_end]

    # ✅ Train XGBoost model
    xgb_model = xgb.XGBRegressor(eval_metric='mae', early_stopping_rounds=8)
    xgb_model.fit(X_train_cv, y_train_cv, eval_set=[(X_test_cv, y_test_cv)], verbose=False)

    # ✅ Predict on validation set
    y_pred_cv = xgb_model.predict(X_test_cv)

    # ✅ Compute MAE
    mae = mean_absolute_error(y_test_cv, y_pred_cv)
    mae_scores.append(mae)

    print(f"Split {i+1}: Train Size = {train_end}, Test Size = {test_size}, MAE = {mae:.3f}")

    # ✅ Store the best model using `.save_model()`
    if mae < best_mae:
        best_mae = mae
        xgb_model.get_booster().save_model(best_model_path)  # ✅ Save best model to file

# ✅ Compute average MAE
print("Average MAE from Expanding Window Cross Validation:", np.mean(mae_scores))

# ✅ Load the best model
best_xgb_model = xgb.Booster()
best_xgb_model.load_model(best_model_path)  # ✅ Load from saved file

# ✅ Convert X_test_xgb into XGBoost's DMatrix for efficient predictions
dtest = xgb.DMatrix(X_test_xgb)
y_pred_test = best_xgb_model.predict(dtest)

# ✅ Evaluate MAE on Test Set
final_mae = mean_absolute_error(y_test, y_pred_test)
print(f"Final Test Set MAE (Best Model): {final_mae:.3f}")
```

#### LSTM-Attention-XGBoost 

#### Conventional Way: Have training set and validation set during training and apply model to the test set to evaluate model performance

```python
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import LSTM, BatchNormalization, Dropout, Dense, Flatten, Input, Masking, dot
from tensorflow.keras.optimizers import Adam
import numpy as np
from xgboost import XGBRegressor

# Define a custom cyclical learning rate schedule
class CyclicLearningRate(tf.keras.optimizers.schedules.LearningRateSchedule):
    def __init__(self, initial_learning_rate, maximal_learning_rate, step_size, scale_fn):
        # Convert parameters to tensors with float32 type
        self.initial_learning_rate = tf.convert_to_tensor(initial_learning_rate, dtype=tf.float32)
        self.maximal_learning_rate = tf.convert_to_tensor(maximal_learning_rate, dtype=tf.float32)
        self.step_size = tf.cast(step_size, tf.float32)
        self.scale_fn = scale_fn

    def __call__(self, step):
        step = tf.cast(step, tf.float32)
        # Compute current cycle: cycle = floor(1 + step / (2 * step_size))
        cycle = tf.floor(1 + step / (2 * self.step_size))
        # Compute relative position within the cycle
        x = tf.abs(step / self.step_size - 2 * cycle + 1)
        # Compute the cyclic learning rate
        lr = self.initial_learning_rate + \
             (self.maximal_learning_rate - self.initial_learning_rate) * tf.maximum(0.0, (1 - x)) * self.scale_fn(cycle)
        return lr

    def get_config(self):
        # Return hyperparameters as Python floats (to avoid .numpy() issues)
        return {
            "initial_learning_rate": float(self.initial_learning_rate),
            "maximal_learning_rate": float(self.maximal_learning_rate),
            "step_size": float(self.step_size),
            "scale_fn": self.scale_fn,
        }
    
def scale_function(cycle):
    return 1 / (2 ** (cycle - 1))    


# Set training parameters
epoch = 100
batch_size = 64
hist_size = 24
steps_per_epoch = len(X_train) // batch_size  # Ensure X_train is defined

# Create the cyclic learning rate schedule
cyclic_lr = CyclicLearningRate(
    initial_learning_rate=1e-4,
    maximal_learning_rate=1e-2,
    step_size=6 * steps_per_epoch,
    scale_fn=scale_function
)

# Create the optimizer with the custom cyclic learning rate
optimizer = Adam(learning_rate=cyclic_lr, amsgrad=True)

# Define an early stopping callback
callback = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=8)

# Updated LSTM-Attention Model with Dropout and BatchNormalization
def LSTM_attention():
    inputs = Input(shape=(hist_size, X_train.shape[2]))
    # Mask padding values (zeros)
    masked = Masking(mask_value=0.)(inputs)
    # Use LSTM to capture sequential dynamics, with dropout & recurrent dropout for regularization
    lstm = LSTM(132, return_sequences=True, dropout=0.2, recurrent_dropout=0.2)(masked)
    # Optional: Batch normalization helps stabilize training
    bn = BatchNormalization()(lstm)
    # Compute dot product between LSTM outputs to form a raw attention score matrix
    attention = dot([bn, bn], axes=[2, 2])
    # Convert raw attention scores to weights using a Dense layer with softmax
    attention = Dense(hist_size, activation='softmax')(attention)
    # Use the attention weights to compute a weighted sum of the LSTM outputs (context vector)
    context = dot([attention, bn], axes=[2, 1])
    # Flatten the context vector into a 1D vector
    flattened = Flatten()(context)
    # Final Dense layer outputs the prediction (e.g., price)
    output = Dense(1)(flattened)
    model = Model(inputs=inputs, outputs=output)
    return model

# Build and compile the LSTM-Attention model
LSTM_attention_model = LSTM_attention()
LSTM_attention_model.compile(optimizer=optimizer, loss='mae')
LSTM_attention_model.summary()

print('\nLSTM-Attention is fitting a model on train-validation sets...\n')
LSTM_attention_model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=epoch, 
                          batch_size=batch_size, callbacks=[callback])

print('\n----------------------------------\nLSTM-Attention is predicting...\n')
# Compute residuals: actual minus LSTM-Attention model prediction
residuals_train = y_train - LSTM_attention_model.predict(X_train)
residuals_val   = y_val - LSTM_attention_model.predict(X_val)

print('\n----------------------------------\nXGBoost is fitting a model on residuals...\n')
# Build the XGBoost model for residual modeling with some tuned hyperparameters
xgb_hybrid = XGBRegressor(
    eval_metric='mae', 
    early_stopping_rounds=8,
    max_depth=5,
    learning_rate=0.1,
    n_estimators=1000
)

# Fit the XGBoost model on the residuals (ensure X_train_xgb and X_val_xgb are defined)
history = xgb_hybrid.fit(
    X_train_xgb, 
    residuals_train,
    eval_set=[(X_train_xgb, residuals_train), (X_val_xgb, residuals_val)],
    verbose=False
)

# Combine predictions: LSTM-Attention prediction plus XGBoost residual prediction
y_train_pred = LSTM_attention_model.predict(X_train) + xgb_hybrid.predict(X_train_xgb).reshape(-1, 1)
y_val_pred   = LSTM_attention_model.predict(X_val)   + xgb_hybrid.predict(X_val_xgb).reshape(-1, 1)
```
